```python
import requests                   
from bs4 import BeautifulSoup         #beautifulsoup for web scraping 
import collections
import pandas as pd
import seaborn as sns
```

```python
url='https://en.wikipedia.org/wiki/Machine_learning'
response=requests.get(url)                               
```

```python
text=response.text
```

```python
soup= BeautifulSoup(text, 'html')                #web data in html format
```

```python
soup
```

```python
a=soup.find_all('p')              #searches for all p tag in the html format
list1=[]                          #empty list made
for i in a:
          z=i.text.split()        #splits all words
          for j in z:
            list1.append(j)       #appends all words in the empty list list1
```

```python
len(list1)                        #length of list1
```

```python
print(list1)
```

```python
word_counter = collections.Counter(list1)
for word, count in word_counter.most_common(20):    #here if the arguement is 20 it decides the top most common words to be searched, if blank it searches all the words
    print(word, ": ", count)

print(word_counter.most_common(20))                 #prints top 20 most occured words with their frequency
```

```python
allwords = word_counter.most_common()
df = pd.DataFrame(allwords, columns = ['Word', 'Count'])
df.shape
```

```python
df.head(20).Count.groupby(df.Word).sum().plot(kind='pie', radius=2.5, autopct='%1.1f%%')  #pie chart of top 20 words
```

```python
df.head(20).plot.bar(x='Word',y='Count', figsize=(15,5))     #bar graph for those 20 words
```
