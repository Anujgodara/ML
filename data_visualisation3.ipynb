```
import pandas as pd                #importing pandas for loading data files
import matplotlib.pyplot as plt    #importing matplotlib for ploting of graph
```

```
df=pd.read_csv("bank.csv")         #reading csv file using read_csv
df.head(5)                         #getting top 5 rows of the data file
```

```
credit=df.head(100).CreditScore            #storing top 100 credit score from the file in credit
age=df.head(100).Age                       #storing top 100 age from the file in age
plt.scatter(credit,age)                    #scatter graph between credit and age
plt.grid()                                 #adding grid in the graph
```

```
salary=df.head(100).EstimatedSalary            #storing top 100 data of estimated salary from the file in salary
plt.scatter(age,salary)                        #scatter graph between age and salary
```

```
df.head(100).EstimatedSalary.plot(kind='kde')    #density curve of estimated salary
plt.xlabel("salary")                             #giving label to the x-axis
```

```
df.head(100).Balance.plot(kind='kde')                #density graph of balance
plt.xlabel("balance")                                #labelling x-axis as balance
```

```
total=df.groupby('Geography').NumOfProducts.count()      #grouping according to geography of number of products
countries=['france','spain','germany']                   
plt.plot(countries,total)                                #graph between countries and total
plt.grid()
```

```
total=df.groupby('Gender').NumOfProducts.count()                   #grouping acc to gender of number of products
result=plt.pie(total,labels=('female','male'),shadow=True,autopct='%1.1f%%',explode=(0.05,0.05)) #pie chart
```
