```
import requests
```

```
dir(requests)
```

```
#accessing web using url
webdata=requests.get('https://php.net')
```

```
print(webdata)        #reponse code of http protocol
#200 url is okh
#302 url is redirected
#404 url is not found
#505 url server is not permitting for web scrapping
```

```
# now getting actual data of website
html_data=webdata.text
```

```
from bs4 import BeautifulSoup
```

```
soup=BeautifulSoup(html_data,'html5lib')       #html_data  ,parser=html5lib
```

```
clean_data=soup.get_text()       #finding all data
```

```
print(clean_data)
#saving data permanently
with open('mywebdata.txt','w+') as f:
   f.write(clean_data)
```

```
#reading from a file
f=open('mywebdata.txt','r')
mydata=f.read()
f.close()
```

```
type(mydata)
```

```
#split
import time
for i in mydata.split():
  print(i)
  time.sleep(1)
```

```
newdata=[i for i in mydata.split()]
```

```
#NLTK graphs
import matplotlib.pyplot as plt
import nltk
```

```
dir(nltk)            #FreqDist this will plot data 
```

```
nlpdata=nltk.FreqDist(newdata)
```

```
nlpdata.plot(5)        #top 5 words repeatition
```

```
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```
#removing stopwords
removedata=[i for i in newdata if i.lower() not in stopwords.words('english')]
```

```
print(removedata)
```

```
#apply NTLK FreqDist
nlpremove=nltk.FreqDist(removedata)
```

```
nlpremove.plot(5)
```
